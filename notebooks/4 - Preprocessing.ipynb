```python
# categories = labelencoder.classes_
# encoded_mapping = dict(zip(categories, encoded_location_values))

# # Print the mapping of categories to encoded values
# for category, encoded_value in encoded_mapping.items():
#     print(f"{category} ---- {encoded_value}")
json_path = os.path.join(parent_folder, "data", "json", "row")

print(os.getcwd())
def make_json(encoder, encoded_values, filename: str):
    categories = encoder.classes_
    encoded_mapping = dict(zip(categories, encoded_values))

    for category, encoded_value in encoded_mapping.items():
        print(f"{category} ---- {encoded_value}")
    
    
    file_path = os.path.join(json_path, f"{filename}.json")
    print(file_path)
    with open(file_path, 'w') as f:
        f.write(json.dumps(encoded_mapping))
```
